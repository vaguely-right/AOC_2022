```python
from aocd import get_data
day7 = get_data(year=2022,day=7).split('\n')
from anytree import Node, RenderTree
```

```python
node = '/'
folders = []
files = []
```

```python
for i in day7:
    # Check for a command
    if i[0] == '$':
        # Check for change directory command
        if i[1] == 'cd':
            # Go back to root directory
            if i[2] == '/':
                node = '/'
            # Move up a level
            elif i[2] == '..':
                node = '/'.join(node.split('/')[:-1])
            # Move to a lower directory
            else:
                node = node+'/'+i[2]
                folders.append(node)
    # Not a command, it's a file or folder from a list command
    else:
        # Check if it's a file
        if i[0] != 'dir':
            files.append([node+'/'+i[1],int(i[0])])
```

```python
foldersize = [[f, sum([i[1] for i in files if f in i[0]])] for f in folders]
```

```python
sum([i[1] for i in foldersize if i[1]<=100000])
```

```python
needed = 30000000 - (70000000 - sum([i[1] for i in files]))
needed
```

```python
min([i[1] for i in foldersize if i[1]>=needed])
```
