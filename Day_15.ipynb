```python
from aocd import get_data
import numpy as np
```

```python
day15 = get_data(year=2022,day=15).split('\n')
```

```python
indata = [i.replace(',','').replace(':','').replace('=',' ').split(' ') for i in day15]
sensors = np.array([(int(i[3]),int(i[5])) for i in indata])
beacons = np.array([(int(i[11]),int(i[13])) for i in indata])
distances = np.sum(np.abs(sensors-beacons),axis=1)
```

```python
distances
```

```python
nobeacon = set()
for i in range(len(sensors)):
    if abs(sensors[i,1]-2000000) < distances[i]:
        xmin = sensors[i,0] - round((distances[i] - abs(sensors[i,1]-2000000)))
        xmax = sensors[i,0] + round((distances[i] - abs(sensors[i,1]-2000000)))
        print(sensors[i],distances[i],xmin,xmax)
        for x in range(xmin,xmax):
            nobeacon.add((x,2000000))
            
len(nobeacon)
```

```python
y = -1
foundbeacon = False
while not foundbeacon:
    y += 1
    beacon = set([i for i in range(4000001)])
    for i in range(len(sensors)):
        if abs(sensors[i,1]-y) < distances[i]:
            xmin = sensors[i,0] - round((distances[i] - abs(sensors[i,1]-y)))
            xmax = sensors[i,0] + round((distances[i] - abs(sensors[i,1]-y)))
            for x in range(xmin,xmax):
                beacon.discard(x)
    if len(beacon) > 0:
        foundbeacon = True
    if (y % 100) == 0:
        print(y)
```
