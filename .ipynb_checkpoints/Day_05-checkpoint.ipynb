```python
from aocd import get_data
day5 = get_data(year=2022,day=5).split('\n\n')
```

```python
crates = day5[0].split('\n')
crates
```

```python
instructions = day5[1].split('\n')
instructions
```

```python
# Get the stacks as individual lists for each position
stacks = [[j[i:i+4].strip() for i in range(0,len(j),4)] for j in crates[:8]]
stacks
```

```python
# Pivot the stacks so each list is one stacks
stacks = [[row[i] for row in stacks] for i in range(9)]
stacks
```

```python
# Put the bottom of the stacks in position 0
[i.reverse() for i in stacks]
stacks
```

```python
# Remove the empty spots
for j in range(10):
    [i.remove('') for i in stacks if '' in i]
stacks
```

```python
# Remove the brackets
stacks = [[i[1] for i in j] for j in stacks]
stacks
```

```python
# Split the instructions into usable format
instructions = [i.split(' ')[1:6:2] for i in instructions]
instructions
```

```python
# Convert the instructions from strings to numbers
instructions = [[int(i) for i in j] for j in instructions]
instructions
```

```python
s = stacks.copy()
for i in instructions:
    n = i[0]
    f = i[1] - 1
    t = i[2] - 1
    for j in range(n):
        s[t].append(s[f].pop())
```

```python
''.join([i[-1] for i in s])
```

```python
s = stacks.copy()
for i in instructions:
    n = i[0]
    f = i[1] - 1
    t = i[2] - 1
    [s[t].append(s[f].pop(m)) for m in range(-1*n,0,1)]
```
